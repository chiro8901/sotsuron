```python
import pandas as pd
import numpy as np
data = pd.read_csv("../data/data.csv", sep=",(?=\S)", engine="python")
```

```python
data.head()
```

```python
values = data.followersCount
sorted_values = np.sort(values)[::-1]
n = len(sorted_values)
cdf = np.arange(1, 1+n) / n
```

```python
import matplotlib.pyplot as plt
plt.loglog(sorted_values, cdf)
plt.grid()
plt.xlabel("Follower Count")
plt.ylabel("CDF")
```

```python
mean = sorted_values.mean()
std = sorted_values.std()
median = np.percentile(sorted_values, 50)
q1 = np.percentile(sorted_values, 25)
q3 = np.percentile(sorted_values, 75)
iqr = q3 - q1

print(f"mean:{mean}")
print(f"std:{std}")
print(f"median:{median}")
print(f"iqr:{iqr}")
print(f"len:{len(sorted_values)}")
```
